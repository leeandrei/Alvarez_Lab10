```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
X = np.linspace(-2, 2, 9)
Y = np.linspace(-1, 1, 7)
X_Y = np.column_stack([[x, y] for x in X for y in Y])
plt.figure(figsize=(4, 4))
plt.scatter(X_Y[0], X_Y[1], s=36, color="red", edgecolor="none")

plt.grid(True)
plt.axis("equal")
```

```python
# Rotation
theta = np.pi/3 
B = np.column_stack([[np.cos(theta), np.sin(theta)], [-np.sin(theta), np.cos(theta)]])


X_Y2 = np.dot(B, X_Y)

plt.figure(figsize=(4, 4), facecolor="w")
plt.scatter(X_Y2[0], X_Y2[1], s=36, color="red", edgecolor="none")
plt.grid(True)
plt.axis("equal")
```
